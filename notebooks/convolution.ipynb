Convolutions with simple loops

```python
def conv2d(image, kernel):
    H, W, C_in = image.shape
    K, _, C_in, C_out = kernel.shape

    out = np.zeros((H-K+1, W-K+1, C_out))
    for c_in in range(C_in):
        for c_out in range(C_out):
            for y in range(H-K+1):
                for x in range(W-K+1):
                    for i in range(K):
                        for j in range(K):
                            out[y, x, c_out] += image[y+i, x+j, c_in] * kernel[i,j,c_in,c_out]
    return out
```

```python
image = np.random.randn(32, 32, 3)
kernel = np.random.randn(3, 3, 3, 5)

out = conv2d(image, kernel)
out.shape
```
